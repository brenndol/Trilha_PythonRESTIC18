```python
import pandas as pd
    
maceio_2003 = pd.read_csv('dataset/maceio2003.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
maceio_2013 = pd.read_csv('dataset/maceio2013.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
maceio_2023 = pd.read_csv('dataset/maceio2023.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
```

```python
dataf = pd.concat([maceio_2003, maceio_2013, maceio_2023])
```

```python
dataf ['DATA'] = pd.to_datetime(dataf['DATA (YYYY-MM-DD)'])
dataf.set_index(['DATA'], inplace=True)
dataf.sort_index(inplace=True)
dataf ['Ano'] = dataf.index.year
dataf ['Mes'] = dataf.index.month
dataf ['Dia'] = dataf.index.day
dataf.set_index(['Ano', 'Mes', 'Dia'], inplace=True)
dataf
```

```python
col = []
for column in dataf.columns:
    if "PRESSAO" in column or "PRESSÃO" in column:
        col.append(("Pressão", column))
    elif "TEMPERATURA" in column:
        col.append(("Temperatura", column))
    elif "UMIDADE" in column:
        col.append(("Umidade", column))
    elif "VENTO" in column:
        col.append(("Velocidade do Vento", column))
    else:
        col.append((column, ""))
```

```python
dataf.colunms = pd.MultiIndex.from_tuples(col)
dataf
```

```python
#exercicio 3.1
dataframes = [maceio_2003, maceio_2013, maceio_2023]

results = {}

for year, df in zip([2003, 2013, 2023], dataframes):
    # Filtrar valores não negativos para precipitação e temperatura
    filtered_df = df[(df['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] >= 0) & 
                     (df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'] >= 0)]
    
    # Calcular médias
    mean_precipitation = filtered_df['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].mean()
    mean_temperature = filtered_df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].mean()
    
    # Calcular mínimos
    min_precipitation = filtered_df['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].min()
    min_temperature = filtered_df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].min()
    
    # Calcular máximos
    max_precipitation = filtered_df['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].max()
    max_temperature = filtered_df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].max()
    
    # Armazenar resultados no dicionário
    results[year] = {
        'Média Precipitação (mm)': mean_precipitation,
        'Média Temperatura (°C)': mean_temperature,
        'Mínimo Precipitação (mm)': min_precipitation,
        'Mínimo Temperatura (°C)': min_temperature,
        'Máximo Precipitação (mm)': max_precipitation,
        'Máximo Temperatura (°C)': max_temperature
    }

# Resultados
for year, stats in results.items():
    print(f"Ano: {year}")
    print(stats)
    print()
```

```python
# exercicio 3.3

# Criar uma coluna para a estação do ano
def get_season(month):
    if month in ['Janeiro', 'Fevereiro', 'Março']:
        return 'Verão'
    elif month in ['Abril', 'Maio', 'Junho']:
        return 'Outono'
    elif month in ['Julho', 'Agosto', 'Setembro']:
        return 'Inverno'
    elif month in ['Outubro', 'Novembro', 'Dezembro']:
        return 'Primavera'

maceio_2003['Estação'] = maceio_2003['Mês'].apply(get_season)
maceio_2013['Estação'] = maceio_2013['Mês'].apply(get_season)
maceio_2023['Estação'] = maceio_2023['Mês'].apply(get_season)

# Agrupar por estação e calcular as médias
mean_by_season_2003 = maceio_2003.groupby('Estação').mean()[['Temp. Média (°C)', 'Precipitação (mm)']]
mean_by_season_2013 = maceio_2013.groupby('Estação').mean()[['Temp. Média (°C)', 'Precipitação (mm)']]
mean_by_season_2023 = maceio_2023.groupby('Estação').mean()[['Temp. Média (°C)', 'Precipitação (mm)']]

# Comparar os dados
comparison_seasons = pd.concat([mean_by_season_2003, mean_by_season_2013, mean_by_season_2023], keys=[2003, 2013, 2023])
comparison_seasons.index.names = ['Ano', 'Estação']

print("\nComparação das informações por estação do ano:")
print(comparison_seasons)
```
