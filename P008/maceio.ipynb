```python
import pandas as pd
import time
    
maceio_2003 = pd.read_csv('datasets/maceio2003.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
maceio_2013 = pd.read_csv('datasets/maceio2013.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
maceio_2023 = pd.read_csv('datasets/maceio2023.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
```

```python
maceio_2023.rename(columns={'Data' : 'DATA (YYYY-MM-DD)', 'Hora UTC':'HORA (UTC)'})
```

```python
dataff = pd.concat([maceio_2003, maceio_2013, maceio_2023])
dataff
```

```python
data = dataff["DATA (YYYY-MM-DD)"].values
index = pd.DatetimeIndex(data, name="Data")
dataff.drop("DATA (YYYY-MM-DD)", axis=1, inplace=True)
dataff.set_index(index, inplace=True)
dataff
```

```python
col = []
for column in dataff.columns:
    if "PRESSAO" in column or "PRESSÃO" in column:
        col.append(("Pressão", column))
    elif "TEMPERATURA" in column:
        col.append(("Temperatura", column))
    elif "UMIDADE" in column:
        col.append(("Umidade", column))
    elif "VENTO" in column:
        col.append(("Velocidade do Vento", column))
    else:
        col.append((column, ""))
```

```python
dataff.colunms = pd.MultiIndex.from_tuples(col)
dataff
```

```python
dataff.groupby(dataff.index.year)["PRECIPITAÇÃO TOTAL, HORÁRIO (mm)"].agg(["mean", "min", "max"])
```

```python
dataff.groupby(dataff.index.year)["TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)"].agg(["mean", "min", "max"])
```

```python

meses = dataff.index.get_level_values(0)

estacao = ["Verão" if mes in [1, 2, 3] else "Outono" if mes in [4, 5, 6] else "Inverno" if mes in [7, 8, 9] else "Primavera" for mes in meses]

dataff['Estação'] = estacao
```

```python
dataff
```

```python
hora = dataff["HORA (UTC)"].values
hora = [str(h)[:2] for h in hora if not isinstance(h, float)]
```

```python
intervalo = []

for x in hora:
    if 0 <= int(x) < 6:
        intervalo.append("Madrugada")
    elif (x[1] != 'a') and 6 <= int(x) < 12:
        intervalo.append("Manhã")
    elif (x[1] != 'a') and 12 <= int(x) < 18:
        intervalo.append("Tarde")
    elif (x[1] != 'a') and 18 <= int(x) < 24:
        intervalo.append("Noite")
    else:
        intervalo.append("Outro")
```

```python
if len(intervalo) < len(dataff):
    intervalo = (intervalo * (len(dataff) // len(intervalo))) + intervalo[:len(dataff) % len(intervalo)]
elif len(intervalo) > len(dataff):
    intervalo = intervalo[:len(dataff)]
```

```python
dataff['Intervalo'] = intervalo
```
